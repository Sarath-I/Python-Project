# **Student Management System**

```python
# Student Management System

# Login System
USERNAME = "admin"
PASSWORD = "entri"
attempts = 7

while attempts > 0:
    user = input("Enter username: ")
    paswd = input("Enter password: ")

    if user == USERNAME and paswd == PASSWORD:
        print("\nYou have successfully logged in. Welcome to the Student Management System.\n")
        break
    else:
        attempts -= 1
        print(f"Incorrect password. Attempts left: {attempts}\n")
    if attempts == 0:
        print("Too many attempts failed.")
        exit()

# students.csv file
filename = "students.csv"
try:
    with open(filename, "x", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Student ID", "Roll No", "Name", "Age", "Department", "Marks", "Entry Time"])
except FileExistsError:
    pass

# Auto-generate student ID
def generate_student_id():
    return "SID" + str(random.randint(1000, 9999))

# Add New Student Record
def add_student():
    try:
        student_id = generate_student_id()
        print(f"Generated Student ID: {student_id}")
        roll = int(input("Enter Roll Number: "))
        name = str(input("Enter Name: "))
        age = int(input("Enter Age: "))
        department = input("Enter Department: ")
        marks = int(input("Enter Marks: "))
        entry_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        with open(filename, "a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([student_id, roll, name, age, department, marks, entry_time])

        print("Student added successfully!\n")
    except Exception as e:
        print("Error:", e)

# View All Student Records
def view_students():
    try:
        with open(filename, "r") as file:
            reader = csv.reader(file)
            data = list(reader)
            if len(data) <= 1:
                print("No records found!\n")
                return
            for row in data:
                print(row)
    except Exception as e:
        print("Error reading file:", e)

# Update Student Details
def update_student():
    roll = input("Enter Roll Number to Update Student Details: ")
    updated = False
    rows = []
    try:
        with open(filename, "r") as file:
            reader = csv.reader(file)
            for row in reader:
                if row and row[1] == roll:
                    print("Record found:", row)
                    name = str(input("Enter New Name: "))
                    age = int(input("Enter New Age: "))
                    department = input("Enter New Department: ")
                    marks = int(input("Enter New Marks: "))
                    row = [row[0], roll, name, age, department, marks, datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")]
                    updated = True
                rows.append(row)

        with open(filename, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerows(rows)

        if updated:
            print("Student updated successfully!\n")
        else:
            print("Student not found!\n")
    except Exception as e:
        print("Error:", e)

# Delete Student Record
def delete_student():
    roll = (input("Enter Roll No to delete: "))
    deleted = False
    rows = []
    try:
        with open(filename, "r") as file:
            reader = csv.reader(file)
            for row in reader:
                if row and row[1] != roll:
                    rows.append(row)
                else:
                    deleted = True

        with open(filename, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerows(rows)

        if deleted:
            print("Student deleted successfully!\n")
        else:
            print("Roll number not found!\n")
    except Exception as e:
        print("Error:", e)

# Search student by name or department
def search_student():
    keyword = input("Enter name or department to search: ").lower()
    found = False
    try:
        with open(filename, "r") as file:
            reader = csv.reader(file)
            for row in reader:
                if any(keyword in item.lower() for item in row):
                    print(row)
                    found = True
        if not found:
            print("No matching records found!\n")
    except Exception as e:
        print("Error:", e)

# View top-performing students (based on marks)
def view_top_students():
    try:
        with open(filename, "r") as file:
            reader = csv.reader(file)
            next(reader)
            data = sorted(reader, key=lambda x: float(x[5]), reverse=True)
            print("Top Performing Students:")
            for row in data[:3]:
                print(row)
    except Exception as e:
        print("Error:", e)

# Menu options
while True:
    print("\n MENU Options")
    print("1. Add New Student")
    print("2. View All Students")
    print("3. Update Student Details")
    print("4. Delete Student")
    print("5. Search Student by Name/Department")
    print("6. View Top Performing Students")
    print("7. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        update_student()
    elif choice == "4":
        delete_student()
    elif choice == "5":
        search_student()
    elif choice == "6":
        view_top_students()
    elif choice == "7":
        print("Thank you for using the system. Goodbye!")
        break
    else:
        print("Invalid choice! Please try again.")
```
